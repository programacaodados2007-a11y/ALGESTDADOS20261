```python
#Faça um programa que peça a largura e a altura de um retângulo e calcule sua área.

b = float(input("Qual a base do retângulo? "))
h = float(input("Qual a altura do retângulo? "))
print(f"A área do retângulo é: {(b*h):.2f} cm")
```
