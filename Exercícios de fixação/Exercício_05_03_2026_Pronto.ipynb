```python
#14. Solicite o preço de um produto e o percentual de desconto, converta ambos para float, calcule o valor final com desconto aplicado e exiba o resultado.
preco = float(input('Digite o valor do produto: '))
desconto = float(input('Qual será o desconto aplicado ao produto (em %)? '))
preco = preco - (preco * desconto/100)
print(f'O valor final do seu produto será de R${preco:.2f}')
```
