```python
# 2. Defina uma senha correta no código. Usando while, peça a senha ao usuário repetidamente até que ele acerte.
# A cada tentativa errada, exiba 'Senha incorreta, tente novamente.'.
# Ao acertar, exiba 'Acesso liberado!' e quantas tentativas foram necessárias.

senha_correta = "20022007"
contador = 0

while True:
    senha = input('Digite a senha: ')
    contador += 1
    if senha == senha_correta:
        print(f"Senha correta. Usou {contador} tentativas.")
        break
    else:
        print("Semha incorreta, tente novamnete.")



```

```python
# # 3. Defina um número secreto entre 1 e 100 no código.
# Usando while, leia tentativas do usuário até acertar.
# A cada erro, diga se o número secreto é maior ou menor.
# Ao final, exiba quantas tentativas foram usadas.

num_secreto = 20
contador = 0

while True:
    num = float(input("Doigte um número de 1 a 100: "))
    contador += 1
    if 1 <= num_secreto <= 100:
        if num == num_secreto:
            print(f"VOCÊ ACERTOU! {num_secreto} esse é o número secreto :). E você ultilizou {contador} tentativa (s)")
            break
        else:
            if num < num_secreto:
                print(f"Esse não é o número secreto :( , mas, o número secretro é maior que {num}")
            else:
                print(f"Esse não é o número secreto :( , mas, o número secretro é menor que {num}")
    else:
        print("O número tem que estar entre 1 e 100.")
```

```python
# # 4. Leia um limite positivo.
# Usando while, leia números inteiros positivos do usuário e some-os.
# # Enquanto a soma não ultrapassar o limite, continue pedindo números.
# # Exiba a soma final e a quantidade de números lidos.

limite = 100

soma = 0
contador = 0

while soma <= limite:
    num = int(input("Digite um valor inteiro positivo: "))
    contador += 1
    soma += num
print(f"A soma foi {soma} e estorou na {contador}° tentativa.")
```
