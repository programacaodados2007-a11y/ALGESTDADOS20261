```python
# 4. Leia um número inteiro N positivo. Usando for, calcule e exiba a soma de todos os números de 1 até N. Valide que N é positivo.
n = int(input("Digite um número: "))

if n > 0:
    soma = 0
    for i in range(1,n+1):
        soma+= i
    print(soma)
else:
    print('Número Negativo!')
```
