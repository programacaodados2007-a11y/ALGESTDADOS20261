```python
# 1. Leia a nota de um aluno. Classifique usando elif: nota ≥ 9 → "A — Excelente", ≥ 7 → "B — Aprovado", ≥ 5 → "C — Recuperação", abaixo → "D — Reprovado".
nota = float(input("Digite sua nota: "))

if nota >= 9:
    print("A — Excelente")
elif nota >= 7:
    print("B — Aprovado")
elif nota >= 5:
    print("C — Recuperação")
else:
    print("D — Reprovado")
```

```python
# 2. Leia um número de 1 a 7 e exiba o nome do dia correspondente (1 = Segunda, ..., 7 = Domingo). Para qualquer outro número, exiba "Número inválido".
dia = int(input("Digite um número: "))

if dia == 1:
    print("Domingo")
if dia == 2:
    print("Segunda-Feira")
if dia == 3:
    print("Terça-Feira")
if dia == 4:
    print("Quarta-Feira")
if dia == 5:
    print("Quinta-Feira")
if dia == 6:
    print("Sexta-Feira")
if dia == 7:
    print("Sábado")
```

```python
# 3. Leia uma nota. Valide se está entre 0 e 10 usando encadeamento de comparações (0 <= nota <= 10). Se inválida, exiba "Nota inválida!". Se válida, exiba "Nota registrada: [nota]"
nota = float(input("Qual a sua nota? "))
if 0<= nota <= 10:
    print(f'Nota registrada: {nota}')
else:
    print('Nota inválida.')
```

```python
# 4. Leia a idade. Classifique: menor de 12 → "Criança", 12 a 17 → "Adolescente", 18 a 59 → "Adulto", 60 ou mais → "Idoso". Valide que a idade não é negativa.
idade = int(input("Digie sua idade: "))

if idade < 12:
    print("Criança")
elif 12 <= idade <= 17:
    print("Adolecente")
elif 18 <= idade <= 59:
    print("Adulto")
elif idade >= 60:
    print("Idoso")
else:
    print(f'Idade não valida{idade < 0}')
```
