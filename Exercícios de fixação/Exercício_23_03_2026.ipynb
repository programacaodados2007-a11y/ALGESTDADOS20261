```python
# 13. Leia um número inteiro. Usando elif, classifique em: positivo e par, positivo e ímpar, negativo e par, negativo e ímpar, ou zero. Use o operador % para verificar paridade.

num = int(input("Digite um número inteiro: "))

if num == 0:
    print("Zero")
elif num > 0 and num % 2 == 0:
    print("Positivo e par")
elif num > 0 and num % 2 != 0:
    print("Positivo e ímpar")
elif num < 0 and num % 2 == 0:
    print("Negativo e par")
else:
    print("Negativo e ímpar")
```

```python
# 15. Leia peso (kg) e altura (m). Valide que peso > 0 e 0.5 <= altura <= 2.5. Calcule o IMC. Classifique em: Abaixo do peso (<18.5), Normal (18.5–24.9), Sobrepeso (25–29.9) e Obesidade (≥30). Exiba o IMC com 2 casas decimais e a classificação.
peso = float(input("Digite o peso (kg): "))
altura = float(input("Digite a altura (m): "))

# Validação
if peso <= 0 or not (0.5 <= altura <= 2.5):
    print("Valores inválidos. Verifique peso e altura.")
else:
    # Cálculo do IMC
    imc = peso / (altura ** 2)

    # Classificação
    if imc < 18.5:
        classificacao = "Abaixo do peso"
    elif imc < 25:
        classificacao = "Normal"
    elif imc < 30:
        classificacao = "Sobrepeso"
    else:
        classificacao = "Obesidade"

    # Saída
    print(f"IMC: {imc:.2f}")
    print(f"Classificação: {classificacao}")
```
