```python
# 01 Contagem de Elementos
# Crie uma lista contendo os nomes de 5 vilões clássicos do Mario.
# Utilize a função len() para contar o número total de itens e exiba o resultado no terminal.

viloes = ['Bowser','Bowser Jr.','Koopalings','Wario','King Boo']
contador = len(viloes)
print(f'Aquantidade é de {contador}')
```

```python
# 02 Adição Dinâmica
# Inicie uma lista vazia chamada moedas_coletadas.
# Use um laço de repetição para pedir que o usuário informe 4 valores inteiros e adicione cada um deles ao final da lista usando append().

moedas_coletadas = []

for i in range(4):
    valor = int(input(f"Digte o {i+1}° valor: "))
    moedas_coletadas.append(valor)

print(moedas_coletadas)
```

```python
# 03 Inserção por Posição
# Crie a lista herois = ['Mario', 'Peach'].
# Utilize o método insert() para adicionar o nome 'Luigi' exatamente no meio da lista (índice 1) e exiba a lista final.

herois = ["Mario" , "Peach"]

herois.insert(1 , "Luigi")
#  ou
# herois.insert(len(herois)//2, 'Luigi')

print(herois)
```

```python
# 04 Exclusão por Nome
# Dada a lista viloes = ['Bowser', 'Goomba', 'Wario', 'Koopa'], solicite que o usuário digite um nome.
# Se o nome estiver na lista, use remove() para excluí-lo.

viloes = ['Bowser', 'Goomba', 'Wario', 'Koopa']
nome = input("digite o nome do vilão do Mário: ").capitalize()
if nome in viloes:
    viloes.remove(nome)
    print(viloes)
else:
    print(viloes)

```
