```python
# crie uma tupla chamada numeros = (1, 2, 3, 2, 4, 2, 5) e utilize os métodos próprios de tuplas para
# descobrir quantas vezes o número 2 aparece e em qual posição ele ocorre pela primeira vez;
# imprima essas informações e explique, com base no comportamento da tupla, por que esses
# métodos são úteis mesmo sendo uma estrutura imutável.

numeros = (1, 2, 3, 2, 4, 2, 5)

contador = 0
for i in numeros:
    if numeros[i] == 2:
        contador += 1
        indice = numeros.index(2)
print(f'O primeiro número 2 aparece no índice {indice}')
print(f'O número 2 aparece {contador} vezes')
```
