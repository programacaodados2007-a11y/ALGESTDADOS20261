```python
# 1. Declare duas variáveis a = 15 e b = 4. Calcule e exiba o resultado de todas as operações aritméticas: +, -, *, /, //, % e **.
a = 15
b = 4
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

```python
# 2. Um professor deseja dividir 37 alunos em grupos iguais de 5. Usando operadores aritméticos, descubra:
# quantos grupos completos serão formados
# quantos alunos sobrarão

grupo = 37
print(f'A quantidade de grupos de 5 pessoas serão de {grupo // 5}. E o resto dessa divisão é {grupo % 5}')
```
