```python
# 1. Classificador de Três Números
# Leia três números inteiros.
# Sem usar listas, determine e exiba qual é o maior, qual é o menor e qual é o do meio.
# Use apenas variáveis, operadores relacionais e condicionais aninhadas.
# Considere também o caso em que dois ou mais números sejam iguais.

num1 = int(input("Digite seu primeiro número: "))
num2 = int(input("Digite seu segundo número: "))
num3 = int(input("Digite seu terceiro número: "))

if num1 > num2 and num1 > num2:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num1}.')
elif num2 > num1 and num2 > num3:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num2}.')
elif num3 > num1 and num3 > num2:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num3}.')
elif num1 == num2 == num3:
    print(f'todos os seus números são iguais. Sendo eles {num1}, {num2} e {num3}.')
else:
    print('Valor Inválido.')
```

```python
# 2. Calculadora de Gorjeta
# Leia o valor total de uma conta de restaurante e a nota de avaliação do serviço (1 a 5).
# Valide ambos.
# Calcule a gorjeta: nota 5 → 20%, nota 4 → 15%, nota 3 → 10%, notas 1 e 2 → 0%.
# Exiba o valor da gorjeta e o total a pagar com 2 casas decimais.

valor_conta = float(input('Qual valor total do bens consumidos no estabelecimento? '))
nota_avaliacao = int(input('Dê uma nota de avaliação do nosso serviço entre 1 a 5, (Sendo 1 a nota miníma e 5 a nota máxima): '))

if valor_conta < 0:
    print('Valor inválido')
elif nota_avaliacao < 1 or nota_avaliacao > 5:
    print('Valor inválido')
else:
    match nota_avaliacao:
        case 5:
            percentual = 0.20
        case 4:
            percentual = 0.15
        case 3:
            percentual = 0.10
        case 1 | 2:
            percentual = 0.00

    gorjeta = valor_conta * percentual
    valor_final = valor_conta + gorjeta
    print(f"O valor final de sua conta consiste em: Valor do seu consumo + gorgeta(%), sendo que a gorgeta e relacionada a avaliação de nossos serviços. | Valor final: R$ {valor_final:.2f}.")
```

```python
from IPython.core.inputtransformer2 import ESC_SH_CAP
# 3. Verificador de Ano Bissexto
# Leia um ano.
# Valide que é positivo.
# Use condicionais aninhadas: divisível por 400 → bissexto; senão, divisível por 100 → não-bissexto; senão, divisível por 4 → bissexto; senão → não-bissexto.
# Use variáveis descritivas e parênteses para clareza nas condições.

ano = int(input("Qual ano gostaria de verificar se é Bissexto? "))

if ano > 0:
    if ano % 400 == 0:
        bissexto = True
    else:
        if ano % 100 == 0:
            bissexto = False
        else:
            if ano % 4 == 0:
                bissexto = True
            else:
                bissexto = False
else:
    print(f'O ano {ano} é invalido!')

if bissexto:
    print(f'O ano {ano} é Bissexto!')
else:
    print(f'O ano não é Bissexto!')
```
