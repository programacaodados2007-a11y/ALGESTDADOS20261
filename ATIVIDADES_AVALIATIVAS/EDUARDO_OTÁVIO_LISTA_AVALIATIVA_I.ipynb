```python
# 1. Classificador de Três Números
# Leia três números inteiros.
# Sem usar listas, determine e exiba qual é o maior, qual é o menor e qual é o do meio.
# Use apenas variáveis, operadores relacionais e condicionais aninhadas.
# Considere também o caso em que dois ou mais números sejam iguais.

num1 = int(input("Digite seu primeiro número: "))
num2 = int(input("Digite seu segundo número: "))
num3 = int(input("Digite seu terceiro número: "))

if num1 > num2 and num1 > num2:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num1}.')
elif num2 > num1 and num2 > num3:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num2}.')
elif num3 > num1 and num3 > num2:
    print(f'Dentre os números {num1}, {num2} e {num3}, o maoir entre eles é {num3}.')
elif num1 == num2 == num3:
    print(f'todos os seus números são iguais. Sendo eles {num1}, {num2} e {num3}.')
else:
    print('Valor Inválido.')
```

```python
# 2. Calculadora de Gorjeta
# Leia o valor total de uma conta de restaurante e a nota de avaliação do serviço (1 a 5).
# Valide ambos.
# Calcule a gorjeta: nota 5 → 20%, nota 4 → 15%, nota 3 → 10%, notas 1 e 2 → 0%.
# Exiba o valor da gorjeta e o total a pagar com 2 casas decimais.

valor_conta = float(input('Qual valor total do bens consumidos no estabelecimento? '))
nota_avaliacao = int(input('Dê uma nota de avaliação do nosso serviço entre 1 a 5, (Sendo 1 a nota miníma e 5 a nota máxima): '))

if valor_conta < 0:
    print('Valor inválido')
elif nota_avaliacao < 1 or nota_avaliacao > 5:
    print('Valor inválido')
else:
    match nota_avaliacao:
        case 5:
            percentual = 0.20
        case 4:
            percentual = 0.15
        case 3:
            percentual = 0.10
        case 1 | 2:
            percentual = 0.00

    gorjeta = valor_conta * percentual
    valor_final = valor_conta + gorjeta
    print(f"O valor final de sua conta consiste em: Valor do seu consumo + gorgeta(%), sendo que a gorgeta e relacionada a avaliação de nossos serviços. | Valor final: R$ {valor_final:.2f}.")
```

```python
from IPython.core.inputtransformer2 import ESC_SH_CAP
# 3. Verificador de Ano Bissexto
# Leia um ano.
# Valide que é positivo.
# Use condicionais aninhadas: divisível por 400 → bissexto; senão, divisível por 100 → não-bissexto; senão, divisível por 4 → bissexto; senão → não-bissexto.
# Use variáveis descritivas e parênteses para clareza nas condições.

ano = int(input("Qual ano gostaria de verificar se é Bissexto? "))

if ano > 0:
    if ano % 400 == 0:
        bissexto = True
    else:
        if ano % 100 == 0:
            bissexto = False
        else:
            if ano % 4 == 0:
                bissexto = True
            else:
                bissexto = False
else:
    print(f'O ano {ano} é invalido!')

if bissexto:
    print(f'O ano {ano} é Bissexto!')
else:
    print(f'O ano não é Bissexto!')
```

```python
# 4. Cálculo de Imposto de Renda Simplificado
# Leia o salário bruto mensal.
# Valide que é positivo.
# Calcule o imposto com elif: até R$2.259 → isento (0%); até R$2.826 → 7,5%; até R$3.751 → 15%; até R$4.664 → 22,5%; acima → 27,5%.
# Exiba a alíquota aplicada, o valor do imposto e o salário líquido, todos formatados com 2 casas decimais.

salario_bruto = float(input('Qual seu salário bruto mensal? '))

if salario_bruto < 0:
    print('Não foi posível calcular seu imposto salarial!')
else:
    if salario_bruto <= 2259:
        aliquota = 0.00
    elif salario_bruto <= 2826:
        aliquota = 0.075
    elif salario_bruto <= 3751:
        aliquota = 0,15
    elif salario_bruto <= 4664:
        aliquota = 0.225
    else:
        aliquota = 0.275

aliquota_aplicada = aliquota * salario_bruto
salario_liquido = salario_bruto - aliquota_aplicada

print(f"O valor do imposto a ser descontado do seu salário sera de R${aliquota_aplicada:.2f}, equivalente a {(aliquota * 100):.1f}% do salário bruto mensal. Com isso, seu salário saiu de, R${salario_bruto:.2f} para R${salario_liquido:.2f}.")
```

```python
# 5. Placar de Futebol
# Leia o número de gols do time A e do time B.
# Valide que ambos são zero ou positivos.
# Usando apenas condicionais, determine e exiba: quem venceu (ou se houve empate), a diferença de gols e se a vitória foi por goleada (diferença maior que 3 gols).
# Para empate, exiba a mensagem correspondente.

gols_A = int(input("Quantos gools o time A fez no time B? "))
gols_B = int(input("Quantos gools o time B fez no time A? "))

if gols_A < 0 or gols_B < 0:
    print("Números inválidos, não existe gols negativos!")
else:
    if gols_A > gols_B:
        vencedor = gols_A
        diferenca = gols_A - gols_B
    elif gols_A < gols_B:
        vencedor = gols_B
        diferenca = gols_B - gols_A
    else:
        gols_A == gols_B
        vencedor = 'Empate!'

    if vencedor == 'Empate!':
        print(f'O fim de jogo terminou em {vencedor}')
    elif vencedor == gols_A:
        if diferenca > 3:
            print(f'O time vencedor foi o A, DE GOLEADA!!! Com {gols_A} gols marcados, sobre {gols_B} gols marcados do time B.')
        else:
            print(f'O time vencedor foi o A, com {gols_A} gols marcados.')
    else:
        vencedor == gols_B
        if diferenca > 3:
            print(f'O time vencedor foi o B, DE GOLEADA!!! Com {gols_B} gols marcados, sobre {gols_A} gols marcados do time A.')
        else:
            print(f'O time vencedor foi o B, com {gols_B} gols marcados.')
```

```python
# 6. Validador de Data
# Leia dia, mês e ano separadamente.
# Valide o ano (positivo) e o mês (1 a 12).
# Com if aninhado,
# valide o dia conforme o mês: meses de 30 dias (4, 6, 9, 11), meses de 31 dias (1, 3, 5, 7, 8, 10, 12) e fevereiro (28 dias em anos normais, 29 em bissextos — use a lógica de bissexto aprendida).
# Exiba se a data é válida ou o motivo da invalidade.

dia = int(input('Digite o dia em que se quer verificar: '))
mes = int(input('Digite o mês desejado: '))
ano = int(input('Digite o ano desejado: '))

if ano <= 0:
    print("Ano inválido!")
elif mes < 1 or mes > 12:
    print('Mês inválido')
else:
    bissexto = (ano % 400 == 0) or (ano % 4 == 0 and ano % 100 != 0)

    match mes:
        case 1 | 3 | 5 | 7 | 8 | 10 | 12:
            dias_no_mes = 31
        case 2:
            if bissexto:
                dias_no_mes = 29
            else:
                dias_no_mes = 28
        case _:
            dias_no_mes = 30

    if dia < 1:
        print("Dia inválido.")
    elif dia > dias_no_mes:
        print(f"Dia inválido. O mês {mes} tem no máximo {dias_no_mes} dias.")
    else:
        print('Data válida.')
```
