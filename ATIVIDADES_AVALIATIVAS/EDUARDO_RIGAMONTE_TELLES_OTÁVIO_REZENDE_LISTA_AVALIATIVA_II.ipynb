```python
#01 - Levantamento Salarial
#Uma empresa de médio porte deseja realizar um levantamento sobre a remuneração de seus funcionários.
#Desenvolva um programa que leia os salários de 10 funcionários, um por vez, e, ao final, informe a média salarial da empresa, o maior salário registrado e quantos funcionários recebem acima de R$ 3.000.
#Os valores informados são válidos e positivos.

soma_salarios = 0
maior_salario = 0
funcionarios_acima_3000 = 0
total_funcionarios = 10

for i in range(1, total_funcionarios + 1):
    salario = float(input(f"Digite o salario do {i}° funcionário: R$"))

    soma_salarios += salario

if salario >= 0:
    if salario > maior_salario:
        maior_salario = salario

    if salario > 3000:
        funcionarios_acima_3000 += 1

else:
    print('valor do salario inválido!')

media_salarial = soma_salarios / total_funcionarios

print('')
print(f'A média salarial da empresa é de: R$ {media_salarial}')
print(f'O maior salário registrado foi: R${maior_salario:.2f}')
print(f'Os funcionarios que recebem acima de 3.000: {funcionarios_acima_3000} funcionarios')
```

```python
#2 - Desempenho da Turma
#Em uma instituição de ensino, um professor deseja analisar o desempenho de uma turma sem saber previamente a quantidade de alunos.
#Desenvolva um programa que leia as notas dos alunos até que seja digitado o valor -1 (fim da entrada).
#Ao final, calcule a média da turma, informe quantos alunos foram aprovados (nota ≥ 7) e quantos foram reprovados.

soma = 0
quantidade_alunos = 0
aprovados = 0
reprovados = 0

while True:
    nota = float(input("Digite a nota do aluno ou digite -1 para encerrar: "))

    if nota == -1:
        break

    soma += nota
    quantidade_alunos += 1

    if nota >= 7:
        aprovados += 1
    else:
        reprovados += 1

media = soma / quantidade_alunos

print("")
print(f"A média da turma é de: {media}")
print(f"A quantidade de aprovados é de: {aprovados}")
print(f"A quantidade de reprovados é de: {reprovados}")#
```

```python
#03 - Sistema Bancário de Saques
#Um sistema bancário simples precisa simular operações de saque em uma conta corrente.
#O programa deve iniciar com um saldo informado pelo usuário e permitir saques sucessivos até que seja digitado 0.
#A cada tentativa, verifique se há saldo suficiente, atualize o saldo quando possível e informe ao usuário quando o valor exceder o saldo disponível.

saldo = float(input("Digite o saldo da conta: "))

while True:
    saque = float(input("Digite o valor do saque (0 para encerrar): "))

    if saque == 0:
        break

    if saque <= saldo:
        saldo -= saque
        print(f"Saque realizado com sucesso.")
        print(f"Saldo atual: R$ {saldo}")

    else:
        print("Saldo insuficiente.")

print(f"Saldo final da conta: R$ {saldo}")
```

```python
#04 - Pesquisa Demográfica
#Uma pesquisa demográfica está sendo realizada para identificar o perfil etário de um grupo de pessoas.
#O programa deve ler as idades informadas até que o usuário digite 0, encerrando a coleta.
#Ao final, apresente a média das idades, a maior idade registrada e quantas pessoas possuem mais de 18 anos.

soma_idades = 0
total_pessoas = 0
maior_idade = 0
maiore_18 = 0

while True:
    idade = int(input("Digite a idade da pessoa e depois digite zero(0) para encerrar a quantidade de pessoas. "))

    if idade == 0:
        break

    if idade < 0:
        print('Digite a idade maior que 0!')

    soma_idades += idade
    total_pessoas += 1

    if idade > maior_idade:
        maior_idade = idade

    if idade > 18:
        maiore_18 += 1

if total_pessoas > 0:
    media_idades = soma_idades / total_pessoas

    print('')
    print(f"A média das idades é: {media_idades:.2f}")
    print(f"A maior idade resgistrada foi: {maior_idade}")
    print(f"Quantidade de pessoas com mais de 18 anos foi: {maiore_18}")

else:
    print("Nenhuma idade foi registrada.")
```

```python
#05 - Controle de Acesso
#Um sistema de segurança controla o acesso a um ambiente restrito.
#O programa deve solicitar uma senha numérica e permitir até três tentativas.
#Caso a senha correta (2024) seja informada dentro desse limite, o acesso deve ser liberado.
#Caso contrário, após três tentativas incorretas, o sistema deve bloquear o acesso e informar essa condição ao usuário.

tentativas = 0
senha_correta = 2024

while tentativas < 3:
    senha = int(input("Digite a senha numérica: "))

    if senha == senha_correta:
        print("Acesso liberado!")
        break
    else:
        tentativas = tentativas + 1
        print("Senha incorreta!")

if tentativas == 3 and senha != senha_correta:
    print("Acesso bloqueado!")
```

```python
#06 - Vendas com Desconto
#Uma loja deseja automatizar o cálculo de suas vendas diárias.
#Desenvolva um programa que leia o valor de 8 vendas realizadas.
#Sempre que o valor de uma venda for superior a R$ 100, aplique automaticamente um desconto de 10% sobre esse valor.
#Ao final, informe o total arrecadado após a aplicação dos descontos.

total = 0

for i in range(1,9):
    venda = (float(input(f'Digite o valor da {i}° venda: R$')))

    if venda > 100:
        desconto = venda * 0.10
        venda_desconto = venda - desconto

    total_arrecadado = total + venda_desconto

print('')
print(f"Total arrecadado após descontos: R$ {total_arrecadado}")
```

```python
#07 - Análise de Sequência Numérica
#Um sistema de análise numérica deve processar uma sequência de valores informados pelo usuário.
#O programa deve ler números inteiros até que seja digitado 0.
#Durante a execução, conte quantos números são pares, quantos são ímpares e calcule a soma de todos os números positivos informados.

pares = 0
impares = 0
soma_positivos = 0

numero = int(input("Digite um número: "))

while numero != 0:

    if numero % 2 == 0:
        pares += 1

    else:
        impares += 1

    if numero > 0:
        soma_positivos += numero

    numero = int(input("Digite outro número ou se caso precise parar digite 0: "))

print("")
print(f"Quantidade de pares: {pares}")
print(f"Quantidade de ímpares: {impares:.2f}")
print(f"Soma dos números positivos: {soma_positivos}")
```

```python
#08 - Monitoramento de Peso
#Uma academia deseja monitorar o peso de seus alunos para fins estatísticos.
#O sistema deve ler o peso de 6 pessoas e, ao final, apresentar a média dos pesos, informar quantas pessoas possuem peso superior a 80 kg e indicar qual foi o menor peso registrado.

soma_pesos = 0
acima_80 = 0
menor_peso = 0

for i in range(1,7):
    peso = float(int(input(f'Digite o pessoa da {i}° pessoa: ')))

    soma_pesos += peso

    if peso > 80:
        acima_80 += 1

    if i == 1:
        menor_peso = peso

    else:
        if peso < menor_peso:
            menor_peso = peso

    media = soma_pesos / 6

print('')
print(f'Média dos pesos é: {media:.2f} kg')
print(f'Pessoas acima de 80kg: {acima_80} pessoas')
print(f'Menor peso registrado foi: {menor_peso} kg')
```

```python
#09 - Votação Eletrônica
#Um sistema de votação eletrônica simplificado será utilizado em uma eleição interna.
#Os votos devem ser informados por meio de códigos numéricos (1, 2 ou 3), sendo que o valor 0 indica o encerramento da votação.
#O programa deve contabilizar os votos de cada candidato e, ao final, informar qual candidato obteve a maior quantidade de votos.

candidato1 = 0
candidato2 = 0
candidato3 = 0

voto = int(input("Digite seu voto (1,2, 3 ou 0 para encerrar): "))

while voto != 0:

    if voto == 1:
        candidato1 +=1

    elif voto == 2:
        candidato2 +=1

    elif voto == 3:
        candidato3 +=1

    else:
        print("Essa opção de voto está incorreta, tente novamente!")

    voto = int(input("Digite seu voto (1, 2, 3 ou 0 para encerrar): "))

print("")
print(f"Votos candidato 1: {candidato1}")
print(f"Votos candidato 2: {candidato2}")
print(f"Votos candidato 3: {candidato3}")

print("")
if candidato1 > candidato2 and candidato1 > candidato3:
    print("Candidato 1 venceu!")
elif candidato2 > candidato1 and candidato2 > candidato3:
    print("Candidato 2 venceu!")
elif candidato3 > candidato1 and candidato3 > candidato2:
    print("Candidato 3 venceu!")
else:
    print("Houve empate!")
```

```python
#10 - Produção Semanal
#Uma indústria deseja acompanhar sua produção semanal para avaliar o desempenho operacional.
#O programa deve ler a quantidade produzida em cada um dos 7 dias da semana e, ao final, calcular a média de produção, informar o maior valor registrado e indicar quantos dias tiveram produção inferior a 100 unidades.

soma = 0
maior = 0
inferior_100 = 0

for i in range(1,8):
    producao = int(input(f"Digite a produção do dia {i}: "))

    soma += producao

    if i == 1:
        maior = producao

    else:
        if producao > maior:
            maior = producao

    if producao < 100:
        inferior_100 +=1

media = soma / 7

print("")
print(f"A média de produção é: {media:.2f}")
print(f"A maior produção registrada foi: {maior}")
print(f"Os dias com produção inferior a 100 foi: {inferior_100}")
```

```python
#11 - Múltiplos de Três
#Um sistema matemático deve analisar propriedades de números inteiros informados pelo usuário.
#O programa deve ler valores até que seja digitado -1 e, ao final, informar quantos números são múltiplos de 3 e calcular a média apenas desses números.

quantidade = 0
soma = 0

numero = int(input("Digite um número ou -1 para encerrar: "))

while numero != -1:

    if numero % 3 == 0:
        quantidade +=1
        soma += numero

    numero = int(input("Digite outro número ou -1 para encerrar: "))

if quantidade > 0:
    media = soma / quantidade

else:
    media = 0

print("")
print(f"A quantidade de múltiplos de 3 é: {quantidade}")
print(f"A média dos múltiplos de 3 é: {media:.2f}")
```

```python
# 21. Pesquisa Eleitoral
# Uma pesquisa eleitoral está sendo realizada para identificar a preferência dos eleitores.
# O programa deve ler votos até que seja digitado 0.
# Votos válidos são representados por códigos específicos; quaisquer outros valores devem ser considerados inválidos.
# Ao final, informe a quantidade de votos válidos, inválidos e o candidato vencedor.

votos_vlidos = 0
votos_invalidos = 0
partido_1 = 0
partido_2 = 0
partido_3 = 0

while True:
    voto = int(input("Qual partido você prefere. Caso seja o Partido 'AUMEIO' - DIGITE 1. Caso seje o Partido 'DICUMPRIDO' - DIGITE 2. Caso seje o partido 'CHORORO' - DIGITE 3 (Se deseja parar digite 0): "))

    if voto == 0:
        print("Pesquisa encerrada!")
        break

    if voto != 1 and voto != 2 and voto != 3:
        votos_invalidos = votos_invalidos + 1
        print("Voto inválido.")
    else:
        votos_vlidos += 1
        if voto == 1:
                partido_1 += 1
        elif voto == 2:
                partido_2 += 1
        elif voto == 3:
                partido_3 += 1

        total = votos_vlidos + votos_invalidos

print(f"O total de votos foram de: {total}")
print(f"A quantidade de votos válidos foram de: {votos_vlidos}")
print(f"A quantidade de votos inválidos foram de: {votos_invalidos}")

maior = max(partido_1,partido_2,partido_3)

if maior == partido_1:
    print(f"O partido com maior número de votos foi o AUMEIO")
elif maior == partido_2:
    print(f"O partido com maior número de votos foi o DICUMPRIDO")
else:
    print(f"O partido com maior número de votos foi o CHORORO")

print("Muito obrigado por participar!")
```

```python
# 22. Estacionamento
# Um estacionamento deseja automatizar o cálculo de sua arrecadação diária.
# O programa deve ler o tempo de permanência de cada veículo até que seja digitado 0.
# O valor a ser pago deve ser calculado com base no tempo, com aplicação de desconto para permanências superiores a 5 horas.
# Ao final, apresente o total arrecadado.

total_arrecadado = 0

while True:
    tempo = float(input('Digte qual foi o tempo de permanencia de seu veículo no estacionamento (Em horas, exemplo: 2.5 = 2h30min), após, digite 0 para exibir o valor a ser pago:  '))
    if tempo < 0 or tempo > 23.59:
        print("Tempo Inválido! Verifique novamente.")
        continue
    else:
        if tempo == 0:
            break
        else:
            valor = tempo * 5
            if tempo > 5:
                valor *= 0.9

            print(f"O valor referente a {tempo} horas é de R${valor:.2f}")

            total_arrecadado += valor

print("-"*40)
print(f"O valor total a ser pago será de R${total_arrecadado:.2f} .")
print("-"*40)
```

```python
# 23. Divisíveis por 2 e por 3
# Um sistema deve analisar uma sequência de números inteiros informados pelo usuário até que seja digitado 0. Ao final.
# informe quantos números são divisíveis simultaneamente por 2 e por 3
# além de calcular a média de todos os números informados.

num_simultaneos = 0
soma_de_num = 0
numeros_informados = 0

while True:
    num = int(input("Digite alguns números inteiros, (caso queira parar digite 0): "))
    numeros_informados += 1
    if num == 0:
        break
    else:
        if num % 2 == 0 and num % 3 == 0:
            num_simultaneos += 1
        else:
            print(f"O número {num} não é divisível por 2 nem por 3.")

    soma_de_num += num
    media = soma_de_num / numeros_informados

print("-"*30)
print(f"{num_simultaneos} números são divissíveis por 2 e 3 simultaneamentes.")
print(f"A média entre os números informados é {media}.")
print("-"*30)
```

```python
# 24. Classificação Escolar
# Um sistema escolar precisa classificar o desempenho dos alunos com base em suas médias finais.
# O programa deve ler 8 valores e classificar cada um como aprovado, em recuperação ou reprovado.
# Ao final, apresente a quantidade de alunos em cada categoria.

aprovado = 0
conselho = 0
reprovado = 0

for i in range(8):
    while True:
        media = float(input(f"Digite a mésia do {i+1}° aluno: "))

        if 0 <= media <= 100:
            break
        else:
            print("Valor inválido. A média necessita estar entre 0 e 100.")

    if media >= 59.5:
        aprovado += 1
    elif 50 <= media <= 59.4:
        conselho += 1
    else:
        reprovado += 1
print("-"*30)
print("Resultado final")
print("-"*30)
print(f"Aprovados: {aprovado}")
print(f"Conselho: {conselho}")
print(f"Reprovado: {reprovado}")
print("-"*30)
```

```python
# 25. Análise de Vendas
# Uma empresa deseja analisar o valor de suas vendas.
# O programa deve ler valores até que seja digitado 0.
# Calcular a média das vendas e informar quantas vendas possuem valor superior à média calculada.

quantidade_de_vendas = 0
soma_das_vendas = 0
sup_media = 0

while True:
    vendas = float(input("Digite o valor das vendas em R$: "))

    if vendas == 0:
        break
    else:
        quantidade_de_vendas += 1
        soma_das_vendas += vendas
        media = soma_das_vendas / quantidade_de_vendas

        if vendas > media:
            sup_media += 1

print("-"*30)
print(f"O total de vendas foi de: {quantidade_de_vendas} vendas.")
print("-"*30)
print(f"A média das vendas foi de R${media:.2f}")
print(f"Dentre as {quantidade_de_vendas} vendas efetuadas, {sup_media} foram valores acima da média.")
```

```python
# 26. Perfil Etário
# Um sistema deve coletar a idade de diversas pessoas até que seja digitado 0.
# Ao final, deve calcular a média das idades e informar quantas pessoas são maiores de idade (acima de 18 anos).

idades = 0
soma_das_idades = 0
maior_de_idade = 0

while True:
    idade = int(input("Digite a sua idade: "))

    if idade == 0:
        break
    else:
        idades += 1
        soma_das_idades += idade
        media = soma_das_idades / idades
        if idade >= 18:
            maior_de_idade += 1
print('-'*30)
print(f"A média entre as idades é de {media:.0f} anos.")
print(f"Dentre as {idades} quantidades de idedas informadas, {maior_de_idade} são maiores de idade (de 18 anos acima).")
```

```python
# 27. Par ou Ímpar até N
# Um programa deve ler um número inteiro N e, utilizando repetição, exibir todos os números de 1 até N.
# Para cada número exibido, o sistema deve informar se ele é par ou ímpar.

numero = int(input("Digite um número qualquer: "))

if numero >= 1:
    for i in range(0 , numero + 1):
        if i % 2 == 0:
            print(f"O número {i} é par.")
        else:
            print(f"O número {i} é ímpar.")
else:
    for i in range(1 , numero - 1 , -1):
        if i % 2 == 0:
            print(f"O número {i} é par.")
        else:
            print(f"O número {i} é ímpar.")
```

```python
# 28. Caça aos Primos
# Um sistema deve analisar números inteiros informados pelo usuário até que seja digitado -1.
# Ao final, informe quantos números são primos e calcule a soma dos números que não são primos.

numero_primo = 0
soma_nao_primos = 0

while True:
    numero = int(input("Digite um número inteiro: "))

    if numero == -1:
        break
    elif numero <= 0 and numero != -1:
        print("Não existe números primos negativos.")
    else:
        if numero > 1:
            primo = True
            for i in range(2 , numero):
                if numero % i == 0:
                    primo = False
                    break
        else:
            primo = False

        if primo:
            numero_primo += 1
        else:
            soma_nao_primos += numero

print("-"*30)
print(f"A quantidade de numeros primos dentre essa sequência é de {numero_primo} números.")
print(f"A soma do restante dos número que não são primos é de {soma_nao_primos}.")
```

```python
# 29. Avaliação de Notas
# Um sistema de avaliação deve ler notas até que seja digitado 0, calcular a média geral e informar quantas notas estão acima da média.

quantidade_notas = 0
notas_acima_da_media = 0
soma = 0

while True:
    notas = float(input("Digite suas notas: "))

    if notas == 0:
        break
    else:
        quantidade_notas += 1
        soma += notas
        media = soma / quantidade_notas

        if notas > media:
            notas_acima_da_media += 1
print("-"*30)
print(f"A média geral das notas acima é de {media:.2f}.")
print(f"Dentre as notas acima, a quantidade de notas acima da média é de {notas_acima_da_media} notas.")
```

```python
# 30. Cálculo de Fatorial
# Um programa deve ler um número inteiro positivo e calcular seu fatorial utilizando repetição.
# Após isso, deve verificar se o resultado do fatorial é par ou ímpar e apresentar essa informação ao usuário.

numero = int(input("Digite un número que seja inteiro e positivo: "))

fatorial = 1

for i in range(1 , numero + 1):
    fatorial *= i
print(f"O valor {numero}! é {fatorial}")

if fatorial % 2 == 0:
    print(f"O valor {fatorial} do número {numero}! é PAR.")
else:
    print(f"O valor {fatorial} do número {numero}! é ÍMPAR.")
```
