```python
# 16. Tempo de Atendimento
# Um sistema de atendimento deve registrar o tempo gasto no atendimento de clientes em uma fila.
# O programa deve ler o tempo de atendimento de cada cliente até que seja digitado 0.
# Ao final, informe o número total de clientes atendidos e o tempo médio de atendimento.

tempo_total = 0
quantidade_clientes = 0

while True:
    tempo = float(input("Digte o tempo de atendimeno (Digite 0 patra sair): "))

    if tempo == 0:
        break

    tempo_total += tempo
    quantidade_clientes += 1

if quantidade_clientes > 0:
    media = tempo_total / quantidade_clientes
    print(f"O total de clientes atendidos é igual a: {quantidade_clientes}.")
    print(f"Tempo médio de atendimento {media}.")
else:
    print('Nenhum cliente foi atendido.')
```

```python
# 17. Análise de 10 Números
# Um programa deve analisar uma sequência fixa de 10 números inteiros informados pelo usuário.
# Ao final da leitura, o sistema deve informar quantos números são negativos, calcular a soma dos números positivos e indicar o maior valor informado.

quant_negativos = 0
soma_positivos = 0

numero = int(input("Digte 0 1º número: "))
maior = numero

if numero < 0:
    quant_negativos += 1
elif numero > 0:
    soma_positivos += numero

for i in range(2,11):
    numero = int(input(f'digite o {i}º número: '))

    if numero < 0:
        quant_negativos += 1
    elif numero > 0:
        soma_positivos += numero

    if numero > maior:
        maior = numero

print(f"A quantidade de números negativos é {quant_negativos}.")
print(f"A soma dos múmeros positivos é {soma_positivos}.")
print(f"O maior valor informado foi {maior}.")
```

```python
# 18. Consumo de Energia
# Uma concessionária de energia deseja calcular o valor total arrecadado com base no consumo de seus clientes.
# O programa deve ler o consumo de energia de 6 residências.
# Para consumos superiores a 100 kWh, deve ser aplicada uma tarifa diferenciada.
# Ao final, o sistema deve apresentar o total arrecadado.

total_arrecadado = 0

for i in range(1,7):
    consumo = float(input(f"Qual o valor da {i}ª residência (em kwh): "))

    if consumo > 100:
        valor = consumo * 0.80
    else:
        valor = consumo * 0.50

    total_arrecadado += valor

print(f"Total arrecadado: R${total_arrecadado}")
```
