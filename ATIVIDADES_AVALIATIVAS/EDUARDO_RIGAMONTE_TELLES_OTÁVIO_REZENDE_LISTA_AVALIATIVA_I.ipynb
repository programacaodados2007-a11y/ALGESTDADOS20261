```python
# 2. Calculadora de Gorjeta
# Leia o valor total de uma conta de restaurante e a nota de avaliação do serviço (1 a 5).
# Valide ambos.
# Calcule a gorjeta: nota 5 → 20%, nota 4 → 15%, nota 3 → 10%, notas 1 e 2 → 0%.
# Exiba o valor da gorjeta e o total a pagar com 2 casas decimais.

valor_conta = float(input('Qual valor total do bens consumidos no estabelecimento? '))
nota_avaliacao = int(input('Dê uma nota de avaliação do nosso serviço entre 1 a 5, (Sendo 1 a nota miníma e 5 a nota máxima): '))

if valor_conta < 0:
    print('Valor inválido')
elif nota_avaliacao < 1 or nota_avaliacao > 5:
    print('Valor inválido')
else:
    match nota_avaliacao:
        case 5:
            percentual = 0.20
        case 4:
            percentual = 0.15
        case 3:
            percentual = 0.10
        case 1 | 2:
            percentual = 0.00

    gorjeta = valor_conta * percentual
    valor_final = valor_conta + gorjeta
    print(f"O valor final de sua conta consiste em: Valor do seu consumo + gorgeta(%), sendo que a gorgeta e relacionada a avaliação de nossos serviços. | Valor final: R$ {valor_final:.2f}.")
```

```python
# 6. Validador de Data
# Leia dia, mês e ano separadamente.
# Valide o ano (positivo) e o mês (1 a 12).
# Com if aninhado,
# valide o dia conforme o mês: meses de 30 dias (4, 6, 9, 11), meses de 31 dias (1, 3, 5, 7, 8, 10, 12) e fevereiro (28 dias em anos normais, 29 em bissextos — use a lógica de bissexto aprendida).
# Exiba se a data é válida ou o motivo da invalidade.

dia = int(input('Digite o dia em que se quer verificar: '))
mes = int(input('Digite o mês desejado: '))
ano = int(input('Digite o ano desejado: '))

if ano <= 0:
    print("Ano inválido!")
elif mes < 1 or mes > 12:
    print('Mês inválido')
else:
    bissexto = (ano % 400 == 0) or (ano % 4 == 0 and ano % 100 != 0)

    match mes:
        case 1 | 3 | 5 | 7 | 8 | 10 | 12:
            dias_no_mes = 31
        case 2:
            if bissexto:
                dias_no_mes = 29
            else:
                dias_no_mes = 28
        case _:
            dias_no_mes = 30

    if dia < 1:
        print("Dia inválido.")
    elif dia > dias_no_mes:
        print(f"Dia inválido. O mês {mes} tem no máximo {dias_no_mes} dias.")
    else:
        print('Data válida.')
```

```python
# 9. Conversor de Temperatura Completo
# Leia uma temperatura e a unidade de origem usando match/case ("C" para Celsius, "F" para Fahrenheit ou "K" para Kelvin).
# Valide que temperaturas em Kelvin não sejam negativas.
# Calcule e exiba a temperatura convertida nas outras duas escalas.
# Use variáveis descritivas como temp_celsius, temp_fahrenheit e temp_kelvin.
# Use case _ para unidade inválida.

temperatura = float(input('Digite a temperatura: '))
unidade = input("Digite a unidade de medida (C, F ou K): ").upper()

match unidade:
    case "C":
        temp_celsius = temperatura
        temp_fahrenheit = (temp_celsius * 9/5) + 32
        temp_kelvin = temp_celsius + 273.15

        print(f'Celsius: {temp_celsius:.2f} °C')
        print(f'Fahrenheint: {temp_fahrenheit:.2f} °F')
        print(f'Kelvin: {temp_kelvin:.2f} K')
    case "F":
        temp_fahrenheit = temperatura
        temp_celsius = (temp_fahrenheit - 32) * 5/9
        temp_kelvin = temp_celsius + 273.15

        print(f'Fahrenheint: {temp_fahrenheit:.2f} °F')
        print(f'Celsius: {temp_celsius:.2f} °C')
        print(f'Kelvin: {temp_kelvin:.2f} K')
    case "K":
        if temperatura < 0:
            print("Temeperatura em Kelvim não pode ser negativa.")
        else:
            temp_kelvin = temperatura
            temp_celsius = temp_kelvin - 273.15
            temp_fahrenheit = (temp_celsius * 9/5) + 32

            print(f'Kelvin: {temp_kelvin:.2f} K')
            print(f'Celsius: {temp_celsius:.2f} °C')
            print(f'Fahrenheint: {temp_fahrenheit:.2f} °F')
    case _:
        print('Use uma das unidades.')
```

```python
# 10. Classificador de IMC com Risco
# Leia peso (kg) e altura (m).
# Valide ambos (peso > 0, 0.5 <= altura <= 2.5).
# Calcule o IMC.
# Com elif, classifique em Abaixo do peso, Normal, Sobrepeso ou Obesidade (Grau I, II, III).
# Com if aninhado em cada classificação, informe o risco cardiovascular: Abaixo/Normal → baixo, Sobrepeso/Obesidade I → moderado, Obesidade II/III → alto.
# Exiba o IMC com 2 casas decimais.

peso = float(input("Qual o seu peso (kg)? "))
altura = float(input("Qual a sua altura (m)? "))

if peso <= 0:
    print('Não se existe peso negativo!')
elif altura < 0.5 or altura > 2.5:
    print("Altura inválida")
else:
    imc = peso / (altura ** 2)
    if imc < 18.5:
        classificação = "Magreza"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")
    elif 18.5 <= imc <= 24.9:
        classificação = "Normal"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")
    elif 25 <= imc <= 29.9:
        classificação = "Sobrepeso"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")
    elif 30 <= imc <= 34.9:
        classificação = "Obesidade Grau I"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")
    elif 35 <= imc <= 39.9:
        classificação = "Obesidade Grau II (Severa)"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")
    else:
        classificação = "Obesidade Grau III (Mórbida)"
        print(f"IMC: {imc:.2f}, Classificação: {classificação}.")

    if classificação == "Mgreza" or classificação == "Normal":
        risco = "Baixo"
        print(f"Risco Cardiovascular: {risco}.")
    elif classificação == "Sobrepeso" or classificação == "Obesidade Grau I":
        risco = "Moderado"
        print(f"Risco Cardiovascular: {risco}.")
    else:
        risco = "Alto"
        print(f"Risco Cardiovascular: {risco}.")
```

```python
# 11. Detector de Múltiplos
# Leia um número inteiro positivo.
# Valide a entrada.
# Use o operador % e variáveis booleanas descritivas (eh_multiplo_de_2, eh_multiplo_de_3, eh_multiplo_de_5) para verificar cada divisibilidade.
# Com condicionais, exiba quais múltiplos se aplicam e, ao final, se é múltiplo de todos os três simultaneamente.
# Use parênteses para clareza nas condições compostas.

num = int(input('Digite seu número: '))

if num > 0:
    eh_multiplo_de_2 = (num % 2 == 0)
    eh_multiplo_de_3 = (num % 3 == 0)
    eh_multiplo_de_5 = (num % 5 == 0)

    if eh_multiplo_de_2:
        print(f'O número {num} é multiplo de 2.')
    if eh_multiplo_de_3:
        print(f'O número {num} é multiplo de 3.')
    if eh_multiplo_de_5:
        print(f'O número {num} é multiplo de 5')

    if not (eh_multiplo_de_2 or eh_multiplo_de_3 or eh_multiplo_de_5):
        print(f'O número {num} não é multiplo de 2, 3 e nem 5.')

    if (eh_multiplo_de_2 and eh_multiplo_de_3 and eh_multiplo_de_5):
        print(f'O número {num} é multiplo de 2, 3 e 5.')
else:
    num <= 0
    print('Número inválido.')
```

```python
# 12. Calculadora de Desconto Progressivo
# Leia o preço original de um produto e um código de cupom usando match/case ("PROMO10", "PROMO20", "PROMO30" ou "SEMCUPOM").
# Com if aninhado, verifique se o preço já é menor que R$50,00 — se sim, o desconto máximo permitido é 10%, independente do cupom.
# Exiba o cupom usado, o desconto em reais e o preço final.

preço = float(input('Qual é o preço do seu produto? '))
cupom = input('Digite o cupom: ').upper()

if preço > 0:
    match cupom:
        case "PROMO10":
            desconto = 0.10
        case "PROMO20":
            desconto = 0.20
        case "PROMO30":
            desconto = 0.30
        case _:
            desconto = 0.00

    desconto_valor = preço * desconto
    valor_final = preço - desconto_valor

    if preço < 50:
        desconto == 0.10
        print(f'O desconto maximo desse produto é de 10%. Assim, o desconto adquirido no sue porduto foi de R${desconto_valor:.2f}. Com isso, seu produto sai de R${preço:.2f}, para R${valor_final:.2f}. ')
    elif preço >= 50 and cupom == "PROMO10":
        print(f'Você ultilizou o cupom "PROMO10", com isso, você terá 10% de desconto no seu produto. Desconto: R${desconto_valor:.2f} | Valor inicial = R${preço:.2f} | Valor com desconto = R${valor_final:.2f}.')
    elif preço >= 50 and cupom == "PROMO20":
        print(f'Você ultilizou o cupom "PROMO20", com isso, você terá 20% de desconto no seu produto. Desconto: R${desconto_valor:.2f} | Valor inicial = R${preço:.2f} | Valor com desconto = R${valor_final:.2f}.')
    elif preço >= 50 and cupom == "PROMO30":
        print(f'Você ultilizou o cupom "PROMO30", com isso, você terá 30% de desconto no seu produto. Desconto: R${desconto_valor:.2f} | Valor inicial = R${preço:.2f} | Valor com desconto = R${valor_final:.2f}.')
    else:
        print(f'Você não recebeu nenhum cupom, com isso, não terá desconto. Desconto: R${desconto_valor:.2f} | Valor inicial: R${preço:.2f} | Valor com desconto: R${valor_final:.2f}.')
else:
    print("Valor inválido.")
```

```python
# 13. Avaliador de Triângulo Completo
# Leia os três lados.
# Valide que todos são positivos na camada externa.
# Na camada interna, verifique a desigualdade triangular (cada lado deve ser menor que a soma dos outros dois).
# Se válido, use elif para classificar: equilátero (3 lados iguais), isósceles (exatamente 2 iguais) ou escaleno (todos diferentes).
# Se inválido, exiba mensagem clara.

lado1 = float(input("Digite o tamanha do primeiro lado do triângulo (em cm): "))
lado2 = float(input("Digite o tamanha do segundo lado do triângulo (em cm): "))
lado3 = float(input("Digite o tamanha do terceiro lado do triângulo (em cm): "))

if lado1 > 0 and lado2 > 0 and lado3 > 0:
    if (lado1 > lado2 + lado3) or (lado2 > lado1 + lado3) or (lado3 > lado2 + lado1):
        print("Desigualdade triangular.")
    else:
        if (lado1 == lado2 == lado3):
            print("Triângulo Equilátero.")
        elif (lado1 == lado2) or (lado1 == lado3) or (lado2 == lado3):
            print("Triângulo Isósceles.")
        else:
            (lado1 != lado2 != lado3)
            print("Triângulo Escaleno.")
else:
    print("Valores inválidos.")
```

```python
# 14. Calculadora de Notas com Conceito
# Leia as quatro notas bimestrais de um aluno (0 a 10 cada).
# Valide todas.
# Calcule a média aritmética.
# Use elif para o conceito: média ≥ 9 → A, ≥ 7 → B, ≥ 5 → C, ≥ 3 → D, abaixo → E.
# Com if aninhado, exiba o status: A e B → "Aprovado", C → "Recuperação", D e E → "Reprovado".
# Exiba a média com 2 casas decimais e o conceito.

nota1 = float(input("Digite sua nota do bimestre (de 0 a 10): "))
nota2 = float(input("Digite sua nota do bimestre (de 0 a 10): "))
nota3 = float(input("Digite sua nota do bimestre (de 0 a 10): "))
nota4 = float(input("Digite sua nota do bimestre (de 0 a 10): "))

if (nota1 < 0 or nota2 < 0 or nota3 < 0 or nota4 < 0) or (nota1 > 10 or nota2 > 10 or nota3 > 10 or nota4 > 10):
    print("Não se pode ter notas negativas ou notas maiores que 10.")
else:
    media = (nota1 + nota2 + nota3 + nota4) / 4
    if media >= 9:
        classificação = "a".upper()
        print(f"Classificação da média das notas: {classificação}")
    elif media >= 7:
        classificação = "b".upper()
        print(f"Classificação da média das notas: {classificação}")
    elif media >= 5:
        classificação = "c".upper()
        print(f"Classificação da média das notas: {classificação}")
    elif media >= 3:
        classificação = "d".upper()
        print(f"Classificação da média das notas: {classificação}")
    else:
        classificação = "e".upper()
        print(f"Classificação da média das notas: {classificação}")
```

```python
# 15. Verificador de Acesso por Horário
# Leia a hora atual (0-23) e o tipo de usuário usando match/case ("admin", "operador", "visitante").
# Valide a hora.
# Para cada tipo, use if aninhado com elif para verificar o turno (manhã: 6-11, tarde: 12-17, noite: 18-23, madrugada: 0-5).
# Regras: admin acessa sempre, operador apenas manhã e tarde, visitante apenas tarde.
# Exiba se o acesso é permitido ou negado e o turno atual.


hora = int(input("Digite a hora atual (0-23): "))
usuario = input("Digite o tipo de usuário (admin, operador, visitante): ")

if not (0 <= hora <= 23):
        print("Hora inválida! Por favor, insira um valor entre 0 e 23.")
else:
    if 6 <= hora <= 11:
            turno = "Manhã"
    elif 12 <= hora <= 17:
            turno = "Tarde"
    elif 18 <= hora <= 23:
            turno = "Noite"
    else:
            turno = "Madrugada"

    match usuario:
        case "admin":
            acesso = "PERMITIDO"

        case "operador":
            if turno == "Manhã" or turno == "Tarde":
                acesso = "PERMITIDO"
            else:
                acesso = "NEGADO"

        case "visitante":
            if turno == "Tarde":
                acesso = "PERMITIDO"
            else:
                acesso = "NEGADO"

        case _:
            acesso = "USUÁRIO INVÁLIDO"

    print(f"Turno Atual: {turno}")
    print(f"Resultado: Acesso {acesso}")
```
