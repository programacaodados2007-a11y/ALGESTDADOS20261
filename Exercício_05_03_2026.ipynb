```python
#1. Peça ao usuário que digite sua idade usando input(), converta o valor para inteiro utilizando int() e exiba na tela a idade somada com 5 anos.
idade = int(input("Digite a sua idade: "))
print(idade+5)
```

```python
#2. Solicite ao usuário sua altura em metros, converta o valor digitado para float e mostre o dobro da altura informada.
altura = float(input("Qual a sua altura? "))
print(altura*2)
```

```python
#3. Crie uma variável inteira com valor 100, converta esse valor para str e concatene com a frase " pontos alcançados", exibindo o resultado final.
valor = 100
texto = str(valor)
print(texto + " pontos alcançados")
```

```python
#4. Peça ao usuário que digite 1 para verdadeiro ou 0 para falso, converta primeiro para inteiro e depois para bool, mostrando o valor booleano resultante.
valor = input("Digite 0 ou 1: ")
numero = int(valor)
booleano = bool(numero)
print(booleano)
```

```python
#5. Peça três números ao usuário, converta todos para inteiro e mostre o maior entre eles.
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digte o segundo número: "))
num3 = int(input("Digite o terceiro número: "))
print(max(num1,num2,num3))
```

```python
#6. Peça quatro números decimais ao usuário, converta para float e calcule a média aritmética entre eles.
n1 = float(input("Digite seu primeiro número: "))
n2 = float(input("Digite seu segundo número: "))
n3 = float(input("Digite seu terceiro número: "))
n4 = float(input("Digite seu quarto número: "))
media = (n1+n2+n3+n4)/4
print(media)
```

```python
#7. Peça a idade do usuário em anos, converta para inteiro e calcule aproximadamente quantos meses essa pessoa já viveu.
idade = int(input("Digite sua idade: "))
meses = idade*12
print(meses)
```
